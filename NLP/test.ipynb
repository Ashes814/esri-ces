```python
import spacy
import pandas as pd
from snownlp import SnowNLP
from snownlp import sentiment
```

```python
centuryParkDF = pd.read_csv('世纪公园.csv', encoding='utf-8')
centuryParkDF.head()
```

```python
centuryParkDF = centuryParkDF[['微博正文', '发布位置', '转发数', '评论数', '点赞数', '发布时间']]
centuryParkDF['date'] = pd.to_datetime(centuryParkDF['发布时间'])
centuryParkDF = centuryParkDF.set_index('date', drop=True)
centuryParkDF.head(10)
```

```python
centuryParkDF['not_rent'] = centuryParkDF['微博正文'].apply(lambda row: '租房' not in row)
centuryParkDF.head(10)
```

```python
centuryParkDF[centuryParkDF['not_rent']].info()
```

```python
centuryParkDF = centuryParkDF[centuryParkDF['not_rent']]
```

```python
scores = []
for i in centuryParkDF['微博正文'].values:
    try:
        s = SnowNLP(i)
        scores.append(s.sentiments)

    except:
        scores.append(0.5)

    
```

```python
centuryParkDF['sentment_score'] = scores
```

```python
centuryParkDF.sentment_score.resample('M').mean()
```
